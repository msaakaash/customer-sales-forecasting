### Revenue by segment

```python
import pandas as pd 
import matplotlib.pyplot as plt

rfm = pd.read_csv("../data/processed/rfm_table.csv")
# Example: Revenue by Segment
segment_rev = rfm.groupby('Segment')['Monetary'].sum().sort_values(ascending=False)
segment_rev.plot(kind='bar', title='Revenue by Segment', figsize=(8,5))
plt.ylabel("Total Revenue")
plt.tight_layout()
plt.savefig('../reports/revenue_by_segment.png')
```
